3. Якщо рівняння не має коренів, программа має генерувати помилку за допомогою ключового слова raise, а потім обробляти її.

4. Правильно написана програма при виникненні різних помилок має виводити текст помилки, а також різні повідомлення про те, в чому суть цих помилок. Необроблених помилок бути не повинно.

```
import math

# Усі виключення написані згідно з інформацією на сайті http://www.algebraclass.ru/kvadratnoe-uravnenie/

a = int(input("Enter a:"))
b = int(input("Enter b:"))
c = int(input("Enter c:"))
b = -b



try:
    d = (b ** 2 )- (4 * a * c)

    D = math.sqrt(d)
    

    if a == 0:
        raise ZeroDivisionError("При a = 0, рівняння не має коренів!")



    if (c == 0) and (b != 0) and (a != 0):
        x1 = 0
        x2 = b / a
        print("x1 =", x1)
        print("x2 =", x2)

    if (b == 0) and (c == 0):
        if D == 0:
            x = b / (2 * a)
            print("Це рівняння має лише один корінь, оскільки дискримінант дорівнює 0.")
            print("x =", x)
        else:
            x = 0
            print("Це рівняння має лише один корінь.")
            print("x =", x)

    x1 = (b + D) / (2 * a)
    x2 = (b - D) / (2 * a)

except ValueError as value: 
    print("Неможливо знайти квадратний корінь від'ємного числа. Назва помилки -", value)
except:
    print("Рівняння має 2 корені!")
    print("x1 =", x1)
    print("x2 =", x2)
    

```
